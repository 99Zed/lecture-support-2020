```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
L_0 = 1
n = 0.01
L = [L_0]
T = 100

for t in range(T):
    L = L + [L[t]*np.exp(n)]

solow_df = pd.DataFrame()
solow_df['L'] = L

solow_df.L.plot()

plt.show()
```

```python
E_0 = 1
g = 0.015
E = [E_0]

for t in range(T):
    E = E + [E[t]*np.exp(g)]

solow_df['E'] = E

solow_df.plot()

plt.show()
```

the definition of capital intensity:

> $ \kappa = \frac{K}{Y} $

> $ g_Y = \theta g_{\kappa} + n + g $ 

>(2.1.14) $ \frac{dK}{dt} = sY - \delta K  $

> $ g_{K} =  \frac{1}{K} \frac{dK}{dt} = s\frac{Y}{K} - \delta = \frac{s}{\kappa} - \delta  $

So the proportional rate of growth of capital-intensity $ \kappa $ is:

>(2.2.8) $ g_\kappa = g_K - g_Y =    \left(  \frac{s}{\kappa} - \delta \right)   - \left(  \theta g_{\kappa} + n + g \right) $

> $ (1+\theta) g_\kappa = \frac{s}{\kappa} - \delta - n - g $

>(2.2.9) $ g_\kappa  = \frac{s/\kappa - (n+g+\delta)}{1+\theta} $







```python
κ_0 = 4
κ = [κ_0]
s = 0.05
δ = 0.025
θ = 1

for t in range(T):
    κ = κ + [κ[t]*(1 + (s/κ[t] - (n+g+δ))/(1+θ))]
    
solow_df['κ'] = κ

solow_df.κ.plot()

plt.show()
```

```python
Y = []
K = []
y = []

for t in range(T):
    Y = Y + [(κ[t]**θ)*L[t]*E[t]]
    K = K + [(κ[t]*Y[t])]
    y = y + [Y[t]/L[t]]

Y = Y + [Y[t-1]*np.exp(n+g)]
K = K + [K[t-1]*np.exp(n+g)]
y = y + [y[t-1]*np.exp(g)]

solow_df['Y'] = Y
solow_df['K'] = K
solow_df['y'] = y

solow_df.plot()

plt.show()
```
